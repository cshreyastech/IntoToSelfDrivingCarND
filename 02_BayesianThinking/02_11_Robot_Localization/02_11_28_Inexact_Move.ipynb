```python
# Modify the move function to accommodate the added 
# probabilities of overshooting or undershooting 
# the intended destination.

# p=[0, 1, 0, 0, 0]
p = [0, 0.5, 0, 0.5, 0]
# U = 2
# q = [0.4, 0.05, 0.05, 0.4, 0.1] 


pExact = 0.8
pOvershoot = 0.1
pUndershoot = 0.1


def move(p, U):
    n = len(p)
    q = [0] * n 
    U %= n
    
    for idx in range(n):
        idx_undershoot = (idx + U - 1) % n
        q[idx_undershoot] += p[idx] * pUndershoot
        
        idx_exact = (idx + U) % n
        q[idx_exact] += p[idx] * pExact
        
        idx_overshoot = (idx + U + 1) % n
        q[idx_overshoot] += p[idx] * pOvershoot
        
    return q
# move(p, U)
```
